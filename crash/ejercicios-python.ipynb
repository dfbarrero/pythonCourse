## Ejercicios básicos sobre Python

### Variables y operadores

1. Crea dos variables con números enteros y muestra su suma.

2. Declara una variable con tu nombre y otra con tu edad, luego imprime ambas con una única instrucción.

3. Calcula el área de un rectángulo con base 5 y altura 10.

4. Convierte una temperatura de 25°C a Fahrenheit (F = C * 9/5 + 32).

5. Calcula el resto de dividir 17 entre 5.

6. Declara dos variables booleanas y muestra el resultado de aplicarles el operador ```and```.

7. Calcula la potencia de 2 elevado a 8.

8. Concatena tres cadenas de texto y muestra el resultado.

9. Calcula el promedio de tres números: 15, 20 y 25.

10. Intercambia los valores de dos variables sin usar una tercera variable.

## Condiciones

1. Escribe un programa que determine si un número es positivo, negativo o cero.

2. Verifica si un número es par o impar.

3. Compara dos números e indica cuál es mayor o si son iguales.

4. Verifica si un número está entre 10 y 20 (inclusive).

## Bucles

1. Imprime los números del 1 al 10 usando un bucle for.

2. Calcula la suma de los números del 1 al 100.

3. Imprime la tabla de multiplicar del 7.

4. Solicita números al usuario hasta que ingrese 0, luego muestra la suma total.

5. Cuenta cuántos números impares hay entre 1 y 30.

## Listas

1. Crea una lista con 5 frutas y muestra el tercer elemento.

2. Añade dos frutas más a la lista de frutas.

3. Elimina la última fruta de la lista usando ```remove()```.

4. Ordena una lista de números de menor a mayor.

5. Encuentra el número máximo y mínimo en una lista.

6. Invierte el orden de los elementos de una lista.

7. Cuenta cuántas veces aparece un elemento específico en una lista.

8. Crea una nueva lista con los cuadrados de los números de otra lista usando comprensión de listas.

9. Dada una lista de palabras ["python", "java", "c++", "javascript", "ruby"], crea una nueva lista usando comprensión que contenga solo las palabras que tienen más de 4 letras.

10. Crea una lista de números [12, 7, 5, 19, 3, 8, 15] y recórrela con un bucle usando enumerate() mostrando el índice y el valor de cada elemento.

## Tuplas

1. Crea una tupla con los días de la semana y muestra el primero y el último.

2. Intenta modificar un elemento de una tupla y observa qué sucede.

## Diccionarios

1. Crea un diccionario con información de una persona (nombre, edad, ciudad) y muestra cada valor.

2. Añade una clave "teléfono" al diccionario anterior.

3. Crea un nuevo diccionario con datos personales como el anterior, con unos valores diferentes. Crea una lista con los dos dicconarios y muestra su contenido.

4. Elimina la clave de un diccionario usando ```del```.

5. Verifica si una clave existe en un diccionario.

6. Muestra todas las claves de un diccionario.

7. Muestra todos los valores de un diccionario.

8. Crea un diccionario que almacene nombres de estudiantes como claves y sus notas como valores, luego calcula el promedio.

9. Recorre un diccionario mostrando cada clave y su valor correspondiente.
